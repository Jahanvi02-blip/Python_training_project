```python
#removing columns that have N/A values
dataset=dataset.drop(columns='Status')
dataset
```

```python
dataset=dataset.drop(columns="unnamed1")
dataset
```

```python
#filling the missing values with 0
dataset=dataset.fillna(0)
dataset
```

```python
# Getting Descriptive statistic summery
dataset.describe()
```

```python
#1.calculate count of male and female and who buy products
import seaborn as sns
sns.countplot(x='Gender',data=dataset)
```

```python
#2.calculate total number of orders from top 10 states
top_10=pd.DataFrame(dataset.groupby('State')['Orders'].sum())
top_10.sort_values(by=['Orders'],inplace=True,ascending=False)
top_10[:10]
```

```python
b=top_10[:10]
sns.barplot(x='Orders',y='State',data=b)
```

```python
#4.top 10 most sold products
most_prod=pd.DataFrame(dataset.groupby('Product_Category')['Orders'].sum())
most_prod.sort_values(by='Orders',inplace=True,ascending=False)
most_prod[:10]
```

```python
#plotting a bargraph of the top 10 most sold products
a=most_prod[:10]
sns.barplot(y="Product_Category",x="Orders",data=a)
```

```python
import matplotlib.pyplot as plt

# Set style for plots
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# 1. Gender Distribution
plt.subplot(2, 3, 1)
sns.countplot(data=dataset, x='Gender', palette='Set2')
plt.title('Gender Distribution')

# 2. Age Group Distribution
plt.subplot(2, 3, 2)
sns.countplot(data=dataset, x='Age Group', hue='Gender', palette='Set2')
plt.title('Age Group by Gender')
plt.xticks(rotation=45)

# 3. State-wise Sales
plt.subplot(2, 3, 3)
state_sales = dataset.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=state_sales.values, y=state_sales.index, palette='viridis')
plt.title('Top 10 States by Sales')

# 4. Zone-wise Sales
plt.subplot(2, 3, 4)
zone_sales = dataset.groupby('Zone')['Amount'].sum().sort_values(ascending=False)
sns.barplot(x=zone_sales.index, y=zone_sales.values, palette='coolwarm')
plt.title('Sales by Zone')

# 5. Top 10 Product Categories
plt.subplot(2, 3, 5)
top_products = dataset['Product_Category'].value_counts().head(10)
sns.barplot(x=top_products.values, y=top_products.index, palette='cubehelix')
plt.title('Top 10 Product Categories')

# 6. Occupation-wise Spending
plt.subplot(2, 3, 6)
occupation_sales = dataset.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=occupation_sales.values, y=occupation_sales.index, palette='Spectral')
plt.title('Top 10 Occupations by Spending')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Age vs Amount, colored by Gender

plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='Age', y='Amount', hue='Gender', alpha=0.6, palette='Set1')
plt.title('Scatter Plot: Age vs Purchase Amount by Gender')
plt.xlabel('Age')
plt.ylabel('Purchase Amount')
plt.grid(True)
plt.show()
```

```python
# Recalculate marital status order count with labels
mar_sta = dataset.groupby(dataset['Marital_Status'].map({0: 'Unmarried', 1: 'Married'}))['Orders'].sum()

# Plot as a pie chart
plt.figure(figsize=(6, 6))
plt.pie(mar_sta)
plt.title('Purchase Orders by Marital Status')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.legend(['Unmarried','Married'],loc='upper left')
plt.show()
```
